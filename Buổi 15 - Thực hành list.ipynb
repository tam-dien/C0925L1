### Bài 1

Viết hàm với đầu vào là một mảng, hãy tính và trả về giá trị trung bình của mảng

```python
def tinh_trung_binh(list_a):
    if len(list_a) == 0:
        return 0
    
    tong = 0
    for so in list_a:
        tong += so

    return tong/len(list_a)

print(tinh_trung_binh([1, 10, 19, 20, 15, 5]))
```

### Bài 2

Viết hàm với đầu vào là một mảng, hãy tìm và trả về giá trị âm lớn nhất trong mảng

```python
def am_lon_nhat(list_a):
    am_max = None
    for x in list_a:
        if x < 0:
            if am_max is None or x > am_max:
                am_max = x
    return am_max

print(am_lon_nhat([1, -5, -8, -3, -10, -2]))
print(am_lon_nhat([1, 2, 3, 7, 10, 2]))
```

### Bài 3

Viết hàm với giá trị đầu vào là một mảng, hãy sắp xếp và trả về mảng có độ lớn tăng dần

```python
def sap_xep(mang):
    if len(mang) <= 1:
        return mang  
    #Sử dụng thuật toán sắp xếp nhanh (Quick Sort)
    
    so_chot = mang[len(mang) // 2]  # chọn phần tử ở giữa làm "chốt" 
    nho_hon = []
    bang = []
    lon_hon = []

    for x in mang:
        if x < so_chot:
            nho_hon.append(x)
        elif x == so_chot:
            bang.append(x)
        else:
            lon_hon.append(x)

    return sap_xep(nho_hon) + bang + sap_xep(lon_hon)

ds = [5, 10, 9, 1, 5, 6, 3, 7, 5, 1, 1, 0]
print("Mảng sau khi sắp xếp tăng dần:", sap_xep(ds))
```

```python
def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n-1):
        for j in range(i + 1, n):
            if arr[j] < arr[i]:
                arr[j], arr[i] = arr[i], arr[j]
    return arr

list_a = [7, 2, 1, 5]
list_b = selection_sort(list_a)

print(list_a)
print(list_b)
```

### Bài 4

Viết hàm với giá trị đầu vào là một mảng, tìm x sao cho đoạn [-x,x] chứa tất cả các giá trị có trong mảng

```python
def find_x(arr):
    if len(arr) == 0:
        return 0
    
    x = abs(arr[0])
    ### x = 1
        
    for i in range(1, len(arr)):
        val = arr[i]
        if abs(val) > x:
            x = abs(val)
    return x


print(find_x([-1, -4, -2]))
```

### Bài 5

Viết hàm với giá trị đầu vào là một mảng, Hãy tìm một vị trí trong mảng thỏa hai điền kiện: có hai giá trị lân cận và giá trị tại vị trí đó bằng tích hai giá trị lân cận. Nếu mảng không tồn tại giá trị như vậy thì hàm trả về giá trị - 1

```python
def tim_vi_tri_thoa_mang(arr):
    for i in range(1, len(arr) - 1):
        if arr[i] == arr[i - 1] * arr[i + 1]:
            return i
    return -1

print(tim_vi_tri_thoa_mang([]))
print(tim_vi_tri_thoa_mang([1]))
print(tim_vi_tri_thoa_mang([5, 7]))
print(tim_vi_tri_thoa_mang([1, 6, 8]))
print(tim_vi_tri_thoa_mang([2, 6, 3]))
print(tim_vi_tri_thoa_mang([4, 6, 12, 2]))
print(tim_vi_tri_thoa_mang([4, 2, 12, 2]))
```

### Bài 6

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng một chiều các số nguyên có chữ sốđầu tiên là chữ số lẻ. Nếu trong mảng không tồn tại giá trị như vậy thì hàm sẽ trả về giá trị 0.

```python
# [123, 758, 694] ~~> 123
# [223, 758, 694] ~~> 758
# [223, 851, 694] ~~> 0

def gia_tri_dau_tien_le(mang):
    for i in range(len(mang)):
        so = mang[i]
        chuoi_so = str(so)
        chu_so_dau = int(chuoi_so[0])
        if chu_so_dau % 2 == 1:
            return so
    return 0

print(gia_tri_dau_tien_le([123, 758, 694]))
print(gia_tri_dau_tien_le([223, 758, 694]))
print(gia_tri_dau_tien_le([223, 851, 694]))
```

### Bài 7

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng có dạng 2^k. Nếu mảng không tồn tại giá trị dạng 2^k thì hàm sẽ trả về giá trị 0.

```python
# 1 2 4 8 16 32
# 1/2 1/4 1/8

def kiem_tra_2_mu_k(a):
    if a < 0:
        return False
    if a < 1:
        a = 1/a

    while a % 2 == 0:
        a //= 2

    return a == 1

def tim_2_mu_2(list_a):
    for so in list_a:
        if kiem_tra_2_mu_k(so):
            return so
        

print(tim_2_mu_2([7, 8, 10, 16])) # 8
print(tim_2_mu_2([1/2, 8, 10, 16])) # 0.5
print(tim_2_mu_2([5, 7, 10, 16])) # 16
```
