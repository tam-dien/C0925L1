### Thêm key vào dictionary

Cú pháp:

`<tên dictionary>[<key>] = <giá trị>`

<b>Ví dụ 1:</b>

```
a = {}
a["tên"] = "NTD"
```

a là một dictionary có một phần tử, phần tử đó có key là "tên" và giá trị của key đó là "NTD"

##### Lưu ý, nếu trong trường hợp đã tồn tại key trong dictionary, dùng lệnh trên sẽ thay đổi giá trị key đó (không phải thêm key khác vào)

<b>Ví dụ 2:</b>

```
a = {}
a["tên"] = "NTD"
a["tên"] = "Nguyễn Tâm Điền"
```

a vẫn là một dictionary có một phần tử, phần tử đó có key là "tên" và giá trị của key đó là "Nguyễn Tâm Điền"

<b>Hãy khai báo một dictionary không có phần tử nào, sau đó thêm một phần tử vào dictionary đó</b>

### Truy xuất phần tử trong dictionary

Cú pháp:

`<tên dictionary>[<key>]`

<b>Ví dụ:</b>

```
a = {"tên": "Nguyễn Tâm Điền", "tuổi": 13, "giới tính": "Nam"}
print(a["tên"])
```

Màn hình sẽ in: Nguyễn Tâm Điền

<b>Hãy khai báo một dictionary rồi in ra màn hình một phần tử bất kỳ trong dictionary đó</b>

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}
score = sinh_vien["score"]

print(score)
````

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}
score = sinh_vien.get("score")

print(score)
````

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}
sinh_vien_2 = sinh_vien

sinh_vien["id"] = 10

print(sinh_vien)
print(sinh_vien_2)
````

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}
sinh_vien_2 = sinh_vien.copy()

sinh_vien["id"] = 10

print(sinh_vien)
print(sinh_vien_2)
````

### Duyệt dictionary bằng lệnh for

Cú pháp:

`for <tên biến> in <tên dictionary>:`

Biến sẽ có giá trị là từng key trong dictionary

<b>Ví dụ:</b>

```
a = {"tên": "Nguyễn Tâm Điền", "tuổi": 13, "giới tính": "Nam"}
for i in a:
    print(i)
```
    
Màn hình sẽ in ra:

tên

tuổi

giới tính

<b>Hãy khai báo một dictionary rồi in ra màn hình toàn bộ giá trị trong dictionary đó</b>

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}

for key in sinh_vien:
    print(key, sinh_vien[key])
````

````python
sinh_vien = {"id": 1, "name": "Nguyễn Văn A", "age": 19, "gender": "Nam"}

for key, value in sinh_vien.items():
    print(key, value)
````

````python
dict_san_pham = {
    "001": "Cơm sườn",
    "002": "Cơm gà",
    "003": "Cơm trứng",
}

def hien_thi_san_pham():
    ### Hiển thị các sản phẩm trong danh sách dict_san_pham
    print("-------------------")
    if len(dict_san_pham) == 0:
        print("Danh sách sản phẩm rỗng")
    for key, value in dict_san_pham.items():
        print(key, ": ", value, sep="")
    print("-------------------")

def them_san_pham_moi():
    ### Yêu cầu người dùng nhập mã sản phẩm mới
    ### Nếu người dùng nhập mã sản phẩm đã tồn tại ~~> yêu cầu nhập lại
    ### Nếu người dùng nhập mã sản phẩm hợp lệ ~~> yêu cầu người dùng nhập tên sản phẩm mới
    ### Thêm sản phẩm vào dict_san_pham
    while True:
        ma_san_pham = input("Nhập vào mã sản phẩm:")
        if ma_san_pham in dict_san_pham:
            print("Mã sản phẩm", ma_san_pham, "đã tồn tại, vui lòng nhập lại")
            continue

        print("Mã sản phẩm:", ma_san_pham)
        ten_san_pham = input("Nhập tên sản phẩm:")
        dict_san_pham[ma_san_pham] = ten_san_pham
        break

def sua_ten_san_pham():
    ### Yêu cấu người dùng nhập mã sản phẩm mới
    ### Nếu người dùng nhập mã sản phẩm không tồn tại ~~> yêu cầu nhập lại
    ### Nếu người dùng nhập mã sản phẩm hợp lệ ~~> yêu cầu người dùng nhập tên sản phẩm mới
    ### Sửa sản phẩm trong dict_san_pham
    while True:
        ma_san_pham = input("Nhập vào mã sản phẩm:")
        if ma_san_pham not in dict_san_pham:
            print("Mã sản phẩm", ma_san_pham, "không tồn tại, vui lòng nhập lại")
            continue

        print("Mã sản phẩm:", ma_san_pham)
        ten_san_pham = input("Nhập tên sản phẩm mới:")
        dict_san_pham[ma_san_pham] = ten_san_pham
        break

def xoa_san_pham():
    ### Yêu cấu người dùng nhập mã sản phẩm mới
    ### Nếu người dùng nhập mã sản phẩm không tồn tại ~~> yêu cầu nhập lại
    ### Nếu người dùng nhập mã sản phẩm hợp lệ ~~> thì xoá sản phẩm trong dict_san_pham
    while True:
        ma_san_pham = input("Nhập vào mã sản phẩm:")
        if ma_san_pham not in dict_san_pham:
            print("Mã sản phẩm", ma_san_pham, "không tồn tại, vui lòng nhập lại")
            continue

        dict_san_pham.pop(ma_san_pham)
        print("Đã xoá sản phẩm", ma_san_pham)
        break

while True:
    print("----------------------\n")
    print("1: Hiển thị danh sách sản phẩm")
    print("2: Thêm sản phẩm")
    print("3: Sửa sản phẩm")
    print("4: Xoá sản phẩm")
    print("5: Thoát")
    print("\n----------------------")
    choice = input("Nhập lựa chọn:")
    if choice == "5":
        break
    if choice == "1":
        hien_thi_san_pham()
    elif choice == "2":
        them_san_pham_moi()
    elif choice == "3":
        sua_ten_san_pham()
    elif choice == "4":
        xoa_san_pham()
    else:
        print("Nhập không hợp lệ! Vui lòng nhập lại!")

print("Đã thoát! Cảm ơn bạn đã sử dụng chương trình!")
````
