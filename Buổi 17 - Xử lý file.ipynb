# Xử lý file

### Mở file

Lệnh mở file:

`f = open(file_path, mode)`

Sau khi xử lý xong, luôn luôn phải đóng file lại để tránh những lỗi không mong muốn.

`f.close()`

Sử dụng with để tự động đóng file:

```
with open(file_path, mode) as f:
    ## xử lý file
```

##### Lưu ý, để xử lý được tiếng việt, ta cần thêm tham số encoding="utf-8"

### Đọc file

Sử dụng mode `r` để đọc file.

Các lệnh đọc file:

#### `f.read()`

Lệnh này trả về giá trị kiểu dữ liệu string.

Lệnh này trả về toàn bộ nội dung file.

````python
f = open("abc.txt", encoding="utf-8")
noi_dung = f.read()
f.close()
print(noi_dung)
````

#### `f.readline()`

Lệnh này trả về giá trị kiểu dữ liệu string.

Lệnh này trả về một dòng dữ liệu trong file.

````python
with open("abc.txt", encoding="utf-8") as f:
    noi_dung = f.readline()

print(noi_dung)
# print('hello')

if noi_dung == "dòng 1":
    print('yes')
````

#### `f.readlines()`

Lệnh này trả về giá trị kiểu dữ liệu list các string.

Lệnh này trả về một list các dòng dữ liệu trong file.

````python
with open("abc.txt", encoding="utf-8") as f:
    noi_dung = f.readlines()

print(noi_dung)
````

### Ghi file

Sử dụng mode `w` để viết file lại từ đầu, hoặc dùng mode `a` để thêm dữ liệu mới vào file.

Lệnh ghi file:

#### `f.write(content)`

````python
# mode để ghi file:
# w: write
# a: append

with open("abc.txt", mode="w", encoding="utf-8") as f:
    f.write("hello")
````

````python
with open("abc.txt", mode="w", encoding="utf-8") as f:
    print('hello')
````

````python
with open("students.txt", "w", encoding="utf-8") as f:
    f.write("Tuấn Nguyễn")
````

````python
with open("students.txt",encoding="utf-8") as f:
    sinh_vien = f.readlines()

print(sinh_vien)

sinh_vien_moi = input("Nhập tên sinh viên:")
sinh_vien.append(sinh_vien_moi + "\n")

with open("students.txt", "w", encoding="utf-8") as f:
    chuoi = ""
    for sv in sinh_vien:
        chuoi += sv
    f.write(chuoi)
````

````python
### nhập vào vị trí sinh viên cần sửa (từ 0 - 4)
### nhập vào tên sinh viên cần sửa

def main():
    with open("students.txt",encoding="utf-8") as f:
        sinh_vien = f.readlines()

    print(sinh_vien)

    vi_tri_sinh_vien = int(input("Nhập vị trí sinh viên:"))

    if not (0 <= vi_tri_sinh_vien < len(sinh_vien)):
        print("Vị trí sinh viên không hợp lệ")
        return

    sinh_vien_moi = input("Nhập tên sinh viên:")

    if len(sinh_vien_moi) < 5:
        print("Độ dài tên sinh viên không phù hợp")
        return

    sinh_vien[vi_tri_sinh_vien] = sinh_vien_moi + "\n"

    with open("students.txt", "w", encoding="utf-8") as f:
        chuoi = ""
        for sv in sinh_vien:
            chuoi += sv
        f.write(chuoi)

main()
````
